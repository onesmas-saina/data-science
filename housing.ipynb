```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/HR_DATA (1).csv")
#data information
df.info()
```

```python
# @title salary vs satisfaction_level

from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(df['salary'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(df, x='satisfaction_level', y='salary', inner='box', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

```python

#scatter plot
#@title average_montly_hours vs time_spend_company

from matplotlib import pyplot as plt
df.plot(kind='scatter', x='average_montly_hours', y='time_spend_company', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title salary

from matplotlib import pyplot as plt
import seaborn as sns
df.groupby('salary').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title satisfaction_level vs last_evaluation

from matplotlib import pyplot as plt
df.plot(kind='scatter', x='satisfaction_level', y='last_evaluation', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title last_evaluation

from matplotlib import pyplot as plt
df['last_evaluation'].plot(kind='hist', bins=20, title='last_evaluation')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title satisfaction_level

from matplotlib import pyplot as plt
df['satisfaction_level'].plot(kind='hist', bins=50, title='satisfaction_level')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
#salary vs satisfaction level in violin
from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(df['salary'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(df, x='satisfaction_level', y='salary', inner='box', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```
