```python
#barplot
sns.barplot(x='ocean_proximity',y='population',data=df)
```

```python
plt.scatter(df['longitude'],df['latitude'],alpha=0.9,c=df['median_house_value'],cmap=plt.get_cmap('jet'))
plt.colorbar()
```

```python
sns.boxplot(data=df,x='ocean_proximity',y='median_house_value')
```

```python
plt.bar(df['housing_median_age'],df['median_house_value'])
plt.xlabel('housing_median_age')
plt.ylabel('median_house_value')
plt.title('Bar Chart relating housing median age and median house value')
plt.show()
```

```python
plt.pie(df['ocean_proximity'].value_counts(),labels=df['ocean_proximity'].value_counts().index,autopct='%1.1f%%')
plt.title('Pie Chart of ocean proximity')
plt.show()

```
